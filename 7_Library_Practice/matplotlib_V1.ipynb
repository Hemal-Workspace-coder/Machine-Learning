```python
#Importing the Matplotlib library
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
```

```python
# adding title, x-axis & y-axis labels
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin(x)')
plt.title('sine wave')
plt.show()
```

```python
#parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('parabola')
plt.show()
```

```python
plt.plot(x,y, 'g--')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
# sin and cos together
x=np.linspace(0,10,100)
plt.plot(x,np.sin(x),'g--')
plt.plot(x,np.cos(x),'r-')
plt.show()
```

Bar plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1,])
language=['Bangla','English','Mathematics','Biology']
people=[100,70,60,90]
ax.bar(language,people)
plt.xlabel('subject')
plt.ylabel('marks')
plt.title('Mark distribution')
plt.show()
```

Pie Chart  


```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
language=['Bangla','English','Mathematics','Biology']
people=[100,70,60,90]
ax.pie(people,labels=language,autopct='%1.2f%%')
plt.show()
```

Scatter Plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='b')
ax.scatter(x,z,color='r')
plt.show()
```

3D Scatter Plot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter3D(x,y,z,c=z,cmap='Greens')
plt.show()
plt.savefig('scatter.jpg')
```
