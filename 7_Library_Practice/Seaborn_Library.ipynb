```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips=sns.load_dataset('tips')
tips.head()
```

```python
#setting a theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x='total_bill',y='tip', hue='smoker', style='smoker', size='size')
```

```python
# load the iris dataset
iris=sns.load_dataset('iris')
iris.head()
```

Scatter plot

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species')  # hue: differentiation will be made based on this
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_width', hue='species')
```

```python
# heatmap
iris1=iris.drop('species', axis=1)
iris1.head()
```

```python
correlation=iris1.corr()
```

```python
sns.heatmap(data=correlation, annot=True, cmap='Blues')
```
